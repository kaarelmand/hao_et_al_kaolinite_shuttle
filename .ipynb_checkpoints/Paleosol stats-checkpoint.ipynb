# Statistics for paleosols -- Hao et al., 2021

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import rcParams
from matplotlib.transforms import blended_transform_factory
from matplotlib.gridspec import GridSpec
import seaborn as sns

sns.set(style='ticks')
rcParams['svg.fonttype'] = 'none'
```

# Wrangle data

Import data.

```python
data = pd.read_excel('paleosol_compilation_cnm_2021-01-25_resaved.xlsx')
data.columns
```

Select needed paleosols.

```python
selected_sols = [
    'Mt. Roe',
    'Cooper Lake',
    'Kuksha',
    'Ville Marie',
    'Hokkalampi',
    'Hekpoort',
    'Drakenstein',
    'Beaverlodge',
    'Flin Flon',
    'Baraboo',
]

data = data[data['Paleosol'].isin(selected_sols)]

data['A/(CNM+A+K*)'].count()
```

Remove unneeded data.

```python
data = data[data["Relative position"] != 'proto']
data = data[data["Relative position"] != 'sill']
data = data[data["Relative position"] != 'cover']

data['A/(CNM+A+K*)'].count()
```

Cut into categorical time bins.

```python
data['Bin'] = pd.cut(
    x=data['Age (Ga)'],
    bins=[1.5, 2.0, 2.4, 3.0],
    labels=['<2.0 Ga', '2.4-2.0 Ga', '>2.4 Ga'],
)
data["Bin"].cat.reorder_categories(['>2.4 Ga', '2.4-2.0 Ga', '<2.0 Ga'], ordered=True, inplace=True)

data.groupby(by='Bin')['A/(CNM+A+K*)'].count()
```

```python
# Order in wanted pattern.
data['Paleosol'] = pd.Categorical(
    data['Paleosol'], categories=selected_sols, ordered=True,
)
data.sort_values('Paleosol', inplace=True)

# Create nice labels
data['paleoname'] = "~" + data['Age (Ga)'].astype(str) + " Ga " + data['Paleosol'].astype(str)
```

# Boxplots

```python
f1 = plt.figure(figsize=(6, 9))
g1 = GridSpec(nrows=2, ncols=4, figure=f1, hspace=1)
ax1 = f1.add_subplot(g1[0, :])
ax2 = f1.add_subplot(g1[1, :])

sns.boxplot(
    data=data,
    ax=ax2,
    x='Bin',
    y='A/(CNM+A+K*)',
    width=0.6,
    palette='vlag',
    whis=[0, 100],
    notch=True,
)

sns.swarmplot(
    data=data,
    ax=ax2,
    x='Bin',
    y='A/(CNM+A+K*)',
    color=".4",
    size=2.5,
)

for i, u in enumerate(data['Bin'].unique()):
    ax2.text(
        x=i, y=1.02, ha='center',
        s="n={}".format(data[data['Bin'] == u]['A/(CNM+A+K*)'].count()),
    )

ax2.set_xlabel("Age bin")
ax2.set_ylabel("Relative kaolinite estimate\nA/(CNM+A+K*)")

# Panel 2 ------------------------------------------------------------------------

# Data
sns.boxplot(
    data=data, ax=ax1, x='paleoname', y='A/(CNM+A+K*)',
    width=0.6, palette='pastel', whis=[0, 100],
)
sns.swarmplot(
    data=data, ax=ax1, x='paleoname', y='A/(CNM+A+K*)',
    color=".4", size=2.5,
)

# Labels
ax1.set_xticklabels(
    ax1.get_xticklabels(), rotation=45, fontdict=dict(ha='right'),
)
ax1.set_ylabel("Relative kaolinite estimate\nA/(CNM+A+K*)")
ax1.set_xlabel("Paleosol")

# Divide by age
for x in (2.5, 3.5, 6.5):
    ax1.axvline(x=x, ls='dotted', color=sns.axes_style()['axes.edgecolor'])
for x, s in zip((1, 5, 8), ('pre-GOE', 'GOE and LE', 'post-LE')):
    ax1.text(
        x=x, y=1.02, s=s, ha='center', va='bottom',
        transform=blended_transform_factory(ax1.transData, ax1.transAxes)
    )
    
# Figure labels
for ax, s in zip((ax1, ax2), ('a', 'b')):
    ax.text(
        transform=ax.transAxes, x=0, y=1.02, s=s,
        fontdict=dict(weight='bold'), ha='left', va='bottom',
    )

f1.align_ylabels()

# Save
for ext in ('svg', 'png'):
    f1.savefig(f'paleosol_boxes.{ext}', bbox_inches='tight',  dpi=350)
```

# P leaching figure

```python
f2, ax21 = plt.subplots()

sns.boxplot(
    data=data, ax=ax21, x='Bin', y='Tau P/Ti',
    width=0.6, palette='vlag', whis=[0, 100], notch=True,
)

sns.stripplot(
    data=data, ax=ax21, x='Bin', y='Tau P/Ti',
    color=".4", size=2.5,
)

for i, u in enumerate(data['Bin'].unique()):
    ax21.text(
        x=i, y=1.02, ha='center', transform=blended_transform_factory(ax21.transData, ax21.transAxes),
        s="n={}".format(data[data['Bin'] == u]['Tau P/Ti'].count()),
    )

ax21.set_xlabel("Age bin")
ax21.set_ylabel("τP$_2$O$_5$")

ax21.set_ylim(top=1.5, bottom=-1.2)

# Save
for ext in ('svg', 'png'):
    f2.savefig(f'paleosol_P_boxes.{ext}', bbox_inches='tight', dpi=350)
```
