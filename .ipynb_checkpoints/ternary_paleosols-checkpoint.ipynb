# CNM-A-K ternary plots of ancient paleosol molar compositions

```python
import datetime
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from ternary_helper import *
```

## Import mineral references

```python
mindata = pd.read_excel('mineral_references.xlsx', index_col=0)
ternary_normalize(mindata, 'CNM', 'Al (mol)', 'K (mol)', suffix='_cnm')
ternary_normalize(mindata, 'CN', 'Al (mol)', 'K (mol)', suffix='_cn')
for c in mindata.columns:
    if '_cnm' in c or '_cn' in c:
        mindata[f'{c}-k*'] = mindata[c]
mindata.rename({'K (mol)_cnm-k*': 'K*_cnm-k*'}, axis='columns', inplace=True)

# REMOVE this if you want to add them back in
mindata.drop(labels=['Illite', 'Phlogopite'], axis='index', inplace=True)

mindata.iloc[:, -13:]
```

## Plot data on ternary

```python
# more specific functions for the paleosol project

def tern_ages(df, age_divisions, axeslist, include_ends=True,
              scattercols=['K*_cnm-k*', 'Al (mol)_cnm-k*', 'CNM_cnm-k*'], agecol='Age (Ga)',
              solcol='Paleosol', legend=True, fancylabel=True, title=True, unit=' Ga',
              **kwargs):
    """
    Divide the *df* DataFrame by numbers given in *age_divisions*, corresponding
    to values in *agecol*, then for each division, paint values in *df[scattercols]*
    onto axes listed in *axeslist*, colouring the data by the column *solcol*.
    
    If *legend=True*, print a legend for every ternary ax, smartly handling the position.
    If *fancylabel=True*, concatenate the value of *agecol*, *solcol* and *unit* for the
    legend labels. If *title=True*, draw a title on the top left corner.
    
    kwargs passed to scatter_tern_color().
    """
    # make a new column for fancier legend handles:
    if fancylabel:
        df = df.copy()
        df['orig_sol'] = df[solcol]
        df[solcol] = df['orig_sol'] + ', ' + df[agecol].map('{:,.2f}'.format) + unit
    
    # cut data and scatter
    # TODO: use pd.cut() instead!
    agesdict = cut_dataframe(df, column=agecol, divisions=age_divisions,
                             include_ends=include_ends, append=unit)
    for tax, (name, subdf) in zip(axeslist, agesdict.items()):
        scatter_tern_color(tax, subdf, scattercols[0], scattercols[1], scattercols[2],
                           colorcol=solcol, **kwargs)
        
        # handle legend; if too big, shift the position outside the axes
        if legend:
            if len(subdf[solcol].unique()) > 6:
                loc = 'upper left'
                bbox_to_anchor = (0.7, 1)
            else:
                loc = 'lower center'
                bbox_to_anchor = None
            tax.legend(fontsize='small', loc=loc, framealpha=1, handlelength=1,
                       handletextpad=0.5, bbox_to_anchor=bbox_to_anchor)
        if title:
            tax.set_title(name, fontsize='x-large', loc='left')
        
    
def baltic_demonstration(tax, df,
                         scattercols=['K*_cnm-k*', 'Al (mol)_cnm-k*', 'CNM_cnm-k*'],
                         colorcol='Relative kaolinite (wt.%)', title='Baltic paleosol',
                         include_ends=False, **kwargs):
    """
    Little wrapper around scatter_tern_color for the common task of plotting a kaolinite
    demo axes. Also includes a title and formatted legend. *kwargs* passed to
    scatter_tern_color()
    """
    scatter_tern_color(tax, df, scattercols[0], scattercols[1], scattercols[2],
                       colorcol=colorcol, include_ends=include_ends, **kwargs)
    
    tax.set_title(title, loc='left', fontsize='x-large')
    tax.legend(title='Relative\nkaolinite (wt.%)', fontsize='small', title_fontsize='small',
               framealpha=1, handlelength=1, handletextpad=0.5, bbox_to_anchor = (0.75, 1),
               loc='upper left')
    tax.get_axes().get_legend().get_title().set_ha('center')
    tax.get_axes().get_legend().get_title().set_position((34, 0))
    

def draw_minerals(tax, mindf, scattercols=['K*_cnm-k*', 'Al (mol)_cnm-k*', 'CNM_cnm-k*'],
                  annotate=True, edgecolor='black', color='black', marker='D', s=30,
                  labelattrs={'Kaolinite':       {'s': 'Kaolinite', 'xy': (47, 85),
                                                  'ha':'right', 'fontsize': 'small'},
                              'Montmorillonite': {'s': 'Smectite', 'xy': (39, 71),
                                                  'ha': 'right', 'fontsize': 'small'},
                              'Orthoclase':      {'s': 'K-feldspar', 'xy': (72, 42),
                                                  'ha': 'right', 'fontsize': 'small'},
                              'Plagioclase':     {'s': 'Plagioclase', 'xy': (20, 38),
                                                  'ha': 'right',  'fontsize': 'small'},
                              'Clinochlore':     {'s': 'Chlorite', 'xy': (26, 16),
                                                  'ha': 'right',  'fontsize': 'small'},
                              'Illite':          {'s': 'Illite', 'xy': (50, 42),
                                                  'xytext': (53, 33),
                                                  'arrowprops': {'arrowstyle': '->'},
                                                  'ha': 'center', 'fontsize': 'small'},
                              'Muscovite':       {'s': 'Muscovite', 'xy': (61 , 68),
                                                  'ha': 'left', 'fontsize': 'small'}},
                  **kwargs):
    """
    Plot mineral data and labels on a ternary ax *tax*.
    """
    for i, mineral in mindata.iterrows():
        tax.scatter([mineral[scattercols].values], edgecolor=edgecolor, color=color,
                    s=s, marker=marker, zorder=3, **kwargs)
    # annotate minerals
    if annotate:
        for mineral in mindf.index:
            if mineral in labelattrs:
                tax.get_axes().annotate(**labelattrs[mineral])
```

Create a CNM-A-K plot per age.

```python
#draw figure
full_cnmk_fig, full_cnmk_axes = draw_ternfig(nrows=3, ncols=2,
                                             labels='cnm-k*', figsize=(11, 15))
for tax in full_cnmk_axes:
    tern_side_scale(tax, tickdict={'40': 70, '60': 80, '80': 90, '100': 100},
                    ylabel='Est. relative kaolinite (%)', bounds=(0, 100), xposition=0,
                    innerticks=True)

# plot data per age
revaxes = list(reversed(full_cnmk_axes))
tern_ages(fulldata, age_divisions=[3.0, 2.65, 2.4, 2.0, 1.0],
          include_ends='low',
          axeslist=revaxes[:-1], s=30, edgecolor='black',
          bring_forward='Hokkalampi, 2.30 Ga')

# plot demonstration data
baltic_demonstration(full_cnmk_axes[0], df=fulldata[fulldata['Paleosol'] == 'Baltic'],
                     s=30, edgecolor='black', colors='plasma')
# plot minerals
draw_minerals(full_cnmk_axes[0], mindata)
```

Comparison of CN-A-K vs CNM-A-K and how it affects the distribution of CIA values.

```python
ternfig_cia, ternaxes_cia = draw_ternfig(ncols=2, labels=('cn-k*', 'cnm-k*'),
                                         figsize=(12, 5))
scatter_tern_color(ternaxes_cia[0], fulldata, left='CN_cn-k*', top='Al (mol)_cn-k*',
                   right='K*_cn-k*', colorcol='CIA (K*)', colors='viridis',
                   divisions=range(40, 101, 10), include_ends=False,
                   edgecolor='black', s=20)
scatter_tern_color(ternaxes_cia[1], fulldata, left='CNM_cnm-k*', top='Al (mol)_cnm-k*',
                   right='K*_cnm-k*', colorcol='CIA (K*)', colors='viridis',
                   divisions=range(40, 101, 10), include_ends=False,
                   edgecolor='black', s=20)
midlegend(ternaxes_cia[0].get_axes(), title='CIA')
```

Draw limited dataset.

```python
# slice data
selected_sols = ['Cooper Lake', 'Kuksha', 'Mt. Roe', 'Hekpoort', 'Ville Marie', 'Drakenstein',
                 'Hokkalampi', 'Beaverlodge', 'Flin Flon', 'Baraboo']
seldata = fulldata[fulldata['Paleosol'].isin(selected_sols)]
```

```python
# draw fig
sel_fig, sel_axes = draw_ternfig(ncols=4, nrows=3, figsize=(22, 15),
                                 labels=('cn', 'cn-k*', 'cnm', 'cnm-k*',
                                         'cn', 'cn-k*', 'cnm', 'cnm-k*',
                                         'cn', 'cn-k*', 'cnm', 'cnm-k*'))

for axlst, sctlst in zip(((sel_axes[8], sel_axes[4], sel_axes[0]),
                          (sel_axes[9], sel_axes[5], sel_axes[1]),
                          (sel_axes[10], sel_axes[6], sel_axes[2]),
                          (sel_axes[11], sel_axes[7], sel_axes[3])),
                         (['K (mol)_cn', 'Al (mol)_cn', 'CN_cn'],
                          ['K*_cn-k*', 'Al (mol)_cn-k*', 'CN_cn-k*'],
                          ['K (mol)_cnm', 'Al (mol)_cnm', 'CNM_cnm'],
                          ['K*_cnm-k*', 'Al (mol)_cnm-k*', 'CNM_cnm-k*'])):
    tern_ages(seldata, age_divisions=(2.4, 2.0), axeslist=axlst,
              scattercols=sctlst, s=30, edgecolor='black',
              legend=False, title=False, bring_forward='Hokkalampi, 2.30 Ga')
for ax, title in zip((sel_axes[1], sel_axes[5], sel_axes[9]),
                     ('>2.4 Ga', '2.4-2.0 Ga', '<2.0 Ga')):
    midlegend(ax.get_axes(), title=title, title_fontsize='x-large')

# draw minerals
minlabels = {'Kaolinite':       {'s': 'Kaolinite\nChlorite', 'xy': (46, 84),
                                 'ha': 'right', 'fontsize': 'small'},
             'Montmorillonite': {'s': 'Smectite', 'xy': (39, 71), 'ha': 'right',
                                 'fontsize': 'small'},
             'Orthoclase':      {'s': 'K-feldspar', 'xy': (72, 42), 'ha': 'right',
                                 'fontsize': 'small'},
             'Plagioclase':     {'s': 'Plagioclase', 'xy': (20, 38), 'ha': 'right',
                                 'fontsize': 'small'},
             'Illite':          {'s': 'Illite', 'xy': (50, 42), 'xytext': (53, 33),
                                 'arrowprops': {'arrowstyle': '->'}, 'ha': 'center',
                                 'fontsize': 'small'},
             'Muscovite':       {'s': 'Muscovite', 'xy': (61 , 68), 'ha': 'left',
                                 'fontsize': 'small'}}

for ax in (sel_axes[0:2]):
    draw_minerals(ax, mindata, scattercols=['K (mol)_cn', 'Al (mol)_cn', 'CN_cn'],
                  color='gold', labelattrs=minlabels)
for ax in (sel_axes[2:4]):
    draw_minerals(ax, mindata, scattercols=['K (mol)_cnm', 'Al (mol)_cnm', 'CNM_cnm'],
                  color='gold')
for ax in sel_axes[4:6] + sel_axes[8:10]:
    draw_minerals(ax, mindata, scattercols=['K (mol)_cn', 'Al (mol)_cn', 'CN_cn'],
                  color='gold', annotate=False)
for ax in sel_axes[6:8] + sel_axes[10:12]:
    draw_minerals(ax, mindata, scattercols=['K (mol)_cnm', 'Al (mol)_cnm', 'CNM_cnm'],
                  color='gold', annotate=False)
```
